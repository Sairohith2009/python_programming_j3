# operators in python

## python have 6 types of operators
1. arithmetic operators
2. assignment operators
3. comparision operators
4. logical operators
5. membership operators
6. identity operators

```python
#arithmetic operators -+,-,*,/,//,%
a=10
b=20
c=30
d=40
print(a+b)
print(b-d)
print(a*c)
print(b/a)
print(b//a)
print(b%a)
```

```python
#comparision operator ->,<,>=,<=, (return true or false)
a=5
b=10
c=16
d=20
print(a>b)
print(b<c)
print(b>=c)
print(b<=c)
print(b==c)
```

```python
# logical operators - and,or,not
# and - this operators checking all conditions are true result is true rest cases failed
# or - this operator check either and conditions true true then result true rest cases failed
a=10
b=20
c=20
d=60
print(a>b and b==c)
print(a<b and c<d)
print(a>b or b==c)
print(a>b or b!=c)
```

```python
#membership operators - checking members are present in that sequence or collection - in , not in 
names=["shiva","vikas","divya","alok"]
print("alok" in names)
print("chandra" in names)
print("Shiva" not in names)
print("Vikas" in names)
```

```python
# identity operators - momery identification of values - is, is not
a=10000
b=10000
print(id(a))
print(id(b))
print(a is b)
```
