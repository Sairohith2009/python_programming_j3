# Iterations in Python
1. iterations are loops used to work on the repreatative tasks or repeatative performances
2. Generally iterations are 2 types
3. `for` loop
4. `while` loop

```python ~ for loop
for iter_variable in sequence:
    print("for is keyword to start the iterations")
    print("iter_variable is variable to iterate data from sequence")
    print("in is keyword to validate the sequence is present or not")
    print("sequence is collection like range,string,list,tuple....etc")
```
```python - while loop
while(condition):
    print("while is keyword to start the loop and validation conditions based on values")
    print("if condition is true then inside statements are executed")

````python
#write a program to print 1 to 60 numbers using for loop?
for num in range (1,61):
     print(num)
````

````python
#write a program to iterate characters from string using for loop?
name="Ankam Sai Rohith"
for char in name:
    print(char)
````

````python
#write a program to print even numbers in between 500 to 600 using for loop?
for n in range(500,601):
    if (n%2 == 0):
        print(n,end=",")
````

````python
#write a program to print odd numbers in between 500 to 600 using for loop?
for n in range(500,601):
    if (n%2 == 1):
        print(n,end=",")
````

````python
#write a program to find the vowels from string using for loop?
name=str(input("please enter your name here:"))
for char in name:
    if(char in "aeiouAEIOU"):
        print(char)
       
        
````

````python
#write a program to find the nonvowels from string using for loop?
name=str(input("please enter your name here:"))
for char in name:
    if(char not in "aeiouAEIOU"):
        print(char)
````

````python
# write a program to iterate names from list of names
names=["Preetham","kalyan","Yaswanth","Ashwin"]
for n in names:
    print("My name is:",n) 
````
